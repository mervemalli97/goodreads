```python
import pandas as pd
import sqlite3
import os
from pandas import Series, DataFrame
```

```python
from glob import glob
```

```python
book_files =sorted(glob('books*.csv'))
review_files =sorted(glob('reviews*.csv'))

book_df = pd.concat((pd.read_csv(file, sep=';', on_bad_lines='skip').assign(filename = file)
          for file in book_files), ignore_index = True)
review_df = pd.concat((pd.read_csv(file, sep=';', on_bad_lines='skip').assign(filename = file)
          for file in review_files), ignore_index = True)
```

```python
book_df = book_df.drop(['Unnamed: 0', 'filename'], axis=1)
book_df = book_df.rename(columns={"0":"NAME", "1":"AUTHOR", "2":"RATING", "3":"RATE_COUNT", "4":"REVIEW_COUNT"})
book_df
```

```python
review_df = review_df.drop(['Unnamed: 0', 'filename'], axis=1)
review_df = review_df.rename(columns={"0":"BOOK_NAME","1":"NAME", "2":"NUMBER", "3":"FOLLOWER_COUNT", "4":"REVIEW", "5":"REVIEW_LIKES", "6":"REVIEW_COMMENTS"})
review_df
```

```python
cnn = sqlite3.connect('goodreads.db')
```

```python
book_df.to_sql('books', cnn, if_exists='replace')
```

```python
review_df.to_sql('reviews', cnn, if_exists='replace')
```

```python
cnn.close()
```
